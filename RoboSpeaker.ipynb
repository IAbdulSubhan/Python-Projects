# Mini-Project- RoboSpeaker

One popular TTS library in Python is gTTS (Google Text-to-Speech).

```python
# pip install gtts
```

```python
from gtts import gTTS  #calling the module for text to speech library
```

```python
from IPython.display import Audio, display   #Library to play with audio
```

```python
def speak(text, language='en', slow=False):
    # Create a text-to-speech object
    tts = gTTS(text=text, lang=language, slow=slow)

    # Save the generated speech to a file
    tts.save("output.mp3")

    # Use IPython display to play the generated speech file
    display(Audio("output.mp3", autoplay=True))
```

```python
#This is just to saying you please enter the text in prompt
welcome = speak("Hi,Happy to see you,  please enter the text in prompt or if you want to exit just type exit, Thankyou")
while True:
    text_to_say = input("Enter the text you want to speak (type 'exit' to stop): ")
    
    if text_to_say.lower() == 'exit':
        break
    
    speak(text_to_say)
    
```
