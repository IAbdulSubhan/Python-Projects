```python
# Read menu content from menu.txt
with open('menu.txt', 'r') as file:
    menu_content = file.read()
```

```python

# Initialize menu dictionary
menu = {}
categories = ["Fast Food", "Ice Cream", "Cold Drinks", "Desi Food"]

# Populate menu dictionary from menu content
current_category = None
for line in menu_content.split('\n'):
    clean_line = line.strip().rstrip(':')  # Remove leading/trailing whitespaces and colons
    if clean_line in categories:
        current_category = clean_line
        menu[current_category] = []
    elif current_category and clean_line:
        menu[current_category].append(clean_line)

```

```python



# Initialize checkout dictionary to store user's order
checkout = {}

def display_menu():
    print("Welcome to Our Restaurant!\n")
    for idx, category in enumerate(categories, start=1):
        print(f"{idx}: {category}")

def display_items(category_index):
    category = categories[category_index - 1]
    items = menu.get(category, [])
    if items:
        print(f"\n{category} Menu:\n")
        for idx, item in enumerate(items, start=1):
            print(f"{idx}. {item}")
    else:
        print("Invalid category selection. Please try again.")

```

```python
def add_to_order(category_index, item_index, quantity):
    category = categories[category_index - 1]
    items = menu.get(category, [])
    if items and 1 <= item_index <= len(items):
        selected_item = items[item_index - 1]
        if category not in checkout:
            checkout[category] = {}
        if selected_item not in checkout[category]:
            checkout[category][selected_item] = 0
        checkout[category][selected_item] += quantity
    else:
        print("Invalid item selection. Please try again.")

```

```python
def display_order():
    print("\nYour Order:\n")
    for category, items in checkout.items():
        print(f"{category}:\n")
        for item, quantity in items.items():
            print(f"{item} - Quantity: {quantity}")
        print("\n")

```

```python
def calculate_total():
    total = 0
    for category, items in checkout.items():
        for item, quantity in items.items():
            price_str = item.split("-")[-1].strip()
            price = int(price_str.split()[-1])
            total += price * quantity
    return total

```

```python
def generate_bill():
    print("\nThank you for dining with us!\n")
    print("*************** Bill *****************\n")
    print("Restaurant: MIAN SONS Restaurant")
    display_order()
    total_amount = calculate_total()
    print(f"Total Amount: PKR {total_amount}\n")
    print("**************************************")

```

```python
# Main program loop
while True:
    display_menu()
    try:
        category_choice = int(input("Select a category by entering its number (1-4): "))
        if 1 <= category_choice <= 4:
            display_items(category_choice)  # This line should be inside the if block
            while True:
                try:
                    item_choice = int(input(f"Select an item by entering its number (1-{len(menu[categories[category_choice-1]])}): "))
                    if 1 <= item_choice <= len(menu[categories[category_choice-1]]):
                        quantity = int(input("Enter the quantity: "))  # Prompt for quantity
                        add_to_order(category_choice, item_choice, quantity)
                        display_order()
                        more_items = input("Do you want to add more items from this category? (yes/no): ").lower()
                        if more_items != 'yes':
                            break
                    else:
                        print(f"Invalid choice. Please enter a number between 1 and {len(menu[categories[category_choice-1]])}.")
                        continue
                except ValueError:
                    print("Invalid input. Please enter a valid number.")
                    continue
            display_order()
            more_categories = input("Do you want to add items from another category? (yes/no): ").lower()
            if more_categories != 'yes':
                break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")
            continue
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        continue

# Ask the user if they are done with the order
done_order = input("Are you done with your order? (yes/no): ").lower()

# Generate and display the bill if the user is done
if done_order == 'yes':
    generate_bill()
else:
    print("Thank you! Feel free to continue adding items.")
```
